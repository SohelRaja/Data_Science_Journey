## Matplotlib Data-Visualization

```python
import matplotlib.pyplot as plt
```

##### To see the plot inside the Jupyter Notebook

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

### Functional method

```python
#functional
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X vs Y')
plt.show()
```

#### Multiplot

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

### Object oriented method(OO)

```python
#Object oriented Method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #[left,botton,width,height] in percentage
axes.plot(x,y)
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('X vs Y')
```

##### Multiple plot in OO

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('X vs Y')

axes2.plot(y,x)
axes2.set_title('Y vs X')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('1st Plot')

axes[1].plot(y,x)
axes[1].set_title('2nd Plot')

plt.tight_layout() #to avoid the overloping of the axis label
```

### Figure size and DPI

```python
fig = plt.figure(figsize=(4,3))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(6,4))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend(loc=0)
```

### Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,color='#24bc44',linewidth=3,alpha=0.5,linestyle='--',marker='o',markersize=20,markerfacecolor='red',
       markeredgewidth=6,markeredgecolor='yellow') #RGB HEX_CODE
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,color='#24bc44',lw=3,ls='--') 
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
