## Seaborn Grid

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
iris = pd.read_csv('Iris.csv')
iris.head()
```

### PAIRPLOT

```python
sb.pairplot(iris)
```

### PairGrid

```python
pg = sb.PairGrid(iris)
pg.map_diag(sb.distplot)
pg.map_upper(plt.scatter)
pg.map_lower(sb.kdeplot)
```

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

### FacetGrid

```python
g = sb.FacetGrid(data=tips,col='time',row='smoker')
g.map(sb.distplot,'total_bill')
```

```python
g = sb.FacetGrid(data=tips,col='time',row='smoker')
g.map(sb.scatterplot,'total_bill','tip')
```
