## Categorical Plots

```python
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
%matplotlib inline
```

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

### Barplot

```python
sb.barplot(x='sex',y='total_bill',data=tips) #x-axis is categorical column and y-axis is numerical column
```

```python
sb.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### Countplot

```python
sb.countplot(x='sex',data=tips)
```

```python
sb.countplot(x='sex',data=tips,hue='sex')
```

### Boxplot

```python
sb.boxplot(x='day',y='total_bill',data=tips)
```

```python
sb.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### Violinplot

```python
sb.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sb.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

### Stripplot

```python
sb.stripplot(x='day',y='total_bill',data=tips)
```

```python
sb.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sb.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sb.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True,jitter=False)
```

### Swarmplot

```python
sb.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sb.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sb.violinplot(x='day',y='total_bill',data=tips)
sb.swarmplot(x='day',y='total_bill',data=tips,hue='sex',color='yellow')
```

### Factorplot

```python
sb.factorplot(x='day',y='total_bill',data=tips)
```

```python
sb.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sb.factorplot(x='day',y='total_bill',data=tips,kind='point')
```

```python
sb.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

### Catplot

```python
sb.catplot(x='day',y='total_bill',data=tips)
```

```python
sb.catplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sb.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
sb.catplot(x='day',y='total_bill',data=tips,kind='bar')
```
