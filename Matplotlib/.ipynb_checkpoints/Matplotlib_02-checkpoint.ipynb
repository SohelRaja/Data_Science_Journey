## Matplotlib Exercies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
%matplotlib inline
```

### Exercise 1

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
axes.plot(x,y)
```

### Exercise 2

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])

axes1.plot(x,y)
axes1.set_title('X vs Y')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')

axes2.plot(y,x)
axes2.set_title('Y vs X')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
```

### Exercise 3

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])

axes1.plot(x,z)
axes1.set_title('X vs Z')
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2.plot(x,y)
axes2.set_title('X vs Y')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
```

### Exercise 4

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

#### Now plot (x,y) and (x,z) on the axes

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,lw=2,ls='--',color='blue')
axes[1].plot(x,z,lw=2,ls='-',color='red')
```
