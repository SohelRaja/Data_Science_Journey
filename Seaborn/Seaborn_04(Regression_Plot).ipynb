## Seaborn Regression_Plots

```python
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
%matplotlib inline
```

```python
tips=pd.read_csv('tips.csv')
tips.head()
```

### LMPLOT

```python
sb.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sb.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','+'])
```

```python
sb.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','+'],scatter_kws={'s':100})
```

```python
sb.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sb.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sb.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.7,size=8)
```
