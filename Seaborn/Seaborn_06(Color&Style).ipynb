## Seaborn Color and Style

```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

```python
sb.set_style('white') #ticks,whitegrid,dark,darkgrid
sb.countplot('sex',data=tips)
sb.despine() #to remove right and top border
```

```python
plt.figure(figsize=(12,3))
sb.countplot('sex',data=tips)
```

```python
sb.set_context('poster')
sb.countplot('sex',data=tips)
```

```python
sb.lmplot('total_bill','tip',data=tips,hue='sex',palette='seismic')
```
